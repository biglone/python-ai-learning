## 1. for 循环

for 循环用于遍历序列（列表、字符串、元组等）中的每个元素。

```python
for 变量 in 序列:
    代码块
```

````python
# 遍历列表
fruits = ["苹果", "香蕉", "橙子", "葡萄"]

for fruit in fruits:
    print(f"我喜欢吃{fruit}")
````

````python
# 遍历字符串
word = "Python"

for char in word:
    print(char, end=" ")
````

### 1.1 range() 函数

range() 生成一个数字序列，常用于 for 循环。

- `range(n)`: 0 到 n-1
- `range(start, stop)`: start 到 stop-1
- `range(start, stop, step)`: 指定步长

````python
# range(n) - 从 0 到 n-1
print("range(5):")
for i in range(5):
    print(i, end=" ")
````

````python
# range(start, stop) - 从 start 到 stop-1
print("range(1, 6):")
for i in range(1, 6):
    print(i, end=" ")
````

````python
# range(start, stop, step) - 指定步长
print("range(0, 10, 2) - 偶数:")
for i in range(0, 10, 2):
    print(i, end=" ")

print("\nrange(10, 0, -1) - 倒数:")
for i in range(10, 0, -1):
    print(i, end=" ")
````

### 1.2 enumerate() - 带索引遍历

````python
fruits = ["苹果", "香蕉", "橙子"]

# 不推荐的写法
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")

print()

# 推荐的写法 - 使用 enumerate
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

print()

# 指定起始索引
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}: {fruit}")
````

### 1.3 zip() - 同时遍历多个序列

````python
names = ["小明", "小红", "小刚"]
ages = [18, 19, 20]
cities = ["北京", "上海", "广州"]

for name, age, city in zip(names, ages, cities):
    print(f"{name}, {age}岁, 来自{city}")
````

## 2. while 循环

while 循环在条件为真时重复执行。

```python
while 条件:
    代码块
```

````python
# 基本 while 循环
count = 0

while count < 5:
    print(f"count = {count}")
    count += 1

print("循环结束")
````

````python
# 计算 1+2+3+...+100
total = 0
n = 1

while n <= 100:
    total += n
    n += 1

print(f"1+2+3+...+100 = {total}")
````

````python
# 猜数字游戏
import random

secret = random.randint(1, 100)
attempts = 0

while True:
    guess = int(input("猜一个1-100的数字: "))
    attempts += 1
    
    if guess < secret:
        print("太小了！")
    elif guess > secret:
        print("太大了！")
    else:
        print(f"恭喜你，猜对了！用了{attempts}次")
        break
````

## 3. break 和 continue

- **break**: 立即退出整个循环
- **continue**: 跳过本次迭代，继续下一次

````python
# break 示例 - 找到第一个偶数就停止
numbers = [1, 3, 5, 6, 7, 8, 9]

for num in numbers:
    if num % 2 == 0:
        print(f"找到第一个偶数: {num}")
        break
    print(f"{num} 是奇数")
````

````python
# continue 示例 - 跳过偶数
for i in range(1, 11):
    if i % 2 == 0:
        continue  # 跳过偶数
    print(i, end=" ")
````

## 4. 循环中的 else

当循环正常完成（没有被 break）时执行 else 块。

````python
# 查找质数
n = 17

for i in range(2, n):
    if n % i == 0:
        print(f"{n} 不是质数，可以被 {i} 整除")
        break
else:
    print(f"{n} 是质数")
````

````python
# 在列表中查找元素
items = ["apple", "banana", "orange"]
target = "grape"

for item in items:
    if item == target:
        print(f"找到了: {target}")
        break
else:
    print(f"没有找到: {target}")
````

## 5. 嵌套循环

````python
# 打印九九乘法表
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f"{j}×{i}={i*j}", end="\t")
    print()  # 换行
````

````python
# 打印三角形
n = 5

for i in range(1, n + 1):
    print("*" * i)
````

````python
# 打印金字塔
n = 5

for i in range(1, n + 1):
    spaces = " " * (n - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)
````

## 6. 实用示例

````python
# 计算阶乘 n!
n = 5
factorial = 1

for i in range(1, n + 1):
    factorial *= i

print(f"{n}! = {factorial}")
````

````python
# 斐波那契数列
n = 10
a, b = 0, 1

print("斐波那契数列:")
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
````

````python
# 统计字符出现次数
text = "hello world"
char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
````
