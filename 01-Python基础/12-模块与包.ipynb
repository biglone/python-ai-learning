## 1. 什么是模块？

模块是包含 Python 代码的 `.py` 文件。模块可以包含函数、类和变量，也可以包含可执行的代码。

## 2. 导入模块

```python
# 方式1：import 模块名
import math

print(f"圆周率: {math.pi}")
print(f"平方根: {math.sqrt(16)}")
print(f"正弦值: {math.sin(math.pi / 2)}")
```

```python
# 方式2：from 模块名 import 成员
from math import pi, sqrt, sin

print(f"圆周率: {pi}")
print(f"平方根: {sqrt(16)}")
```

```python
# 方式3：import 模块名 as 别名
import numpy as np  # 如果安装了 numpy
import math as m

print(f"圆周率: {m.pi}")
```

```python
# 方式4：from 模块名 import *（不推荐）
from math import *

print(f"e = {e}")
print(f"cos(0) = {cos(0)}")
```

## 3. 常用标准库

```python
# os - 操作系统接口
import os

print(f"当前目录: {os.getcwd()}")
print(f"环境变量 PATH: {os.getenv('PATH', 'Not found')[:50]}...")
```

```python
# sys - 系统相关
import sys

print(f"Python 版本: {sys.version}")
print(f"平台: {sys.platform}")
```

```python
# datetime - 日期时间
from datetime import datetime, date, timedelta

now = datetime.now()
print(f"当前时间: {now}")
print(f"格式化: {now.strftime('%Y-%m-%d %H:%M:%S')}")

tomorrow = date.today() + timedelta(days=1)
print(f"明天: {tomorrow}")
```

```python
# random - 随机数
import random

print(f"随机整数 (1-10): {random.randint(1, 10)}")
print(f"随机浮点数 (0-1): {random.random():.4f}")
print(f"随机选择: {random.choice(['苹果', '香蕉', '橙子'])}")

lst = [1, 2, 3, 4, 5]
random.shuffle(lst)
print(f"打乱列表: {lst}")
```

```python
# collections - 高级数据结构
from collections import Counter, defaultdict, namedtuple

# Counter
words = "apple banana apple cherry apple banana".split()
counter = Counter(words)
print(f"词频: {counter}")
print(f"最常见: {counter.most_common(2)}")

# defaultdict
dd = defaultdict(list)
dd["fruits"].append("apple")
dd["fruits"].append("banana")
print(f"defaultdict: {dict(dd)}")

# namedtuple
Point = namedtuple("Point", ["x", "y"])
p = Point(3, 4)
print(f"Point: x={p.x}, y={p.y}")
```

```python
# itertools - 迭代器工具
from itertools import combinations, permutations, product

# 组合
print(f"组合: {list(combinations([1, 2, 3], 2))}")

# 排列
print(f"排列: {list(permutations([1, 2, 3], 2))}")

# 笛卡尔积
print(f"笛卡尔积: {list(product([1, 2], ['a', 'b']))}")
```

```python
# functools - 函数工具
from functools import reduce, lru_cache

# reduce
result = reduce(lambda x, y: x + y, [1, 2, 3, 4, 5])
print(f"reduce 求和: {result}")

# lru_cache - 缓存
@lru_cache(maxsize=100)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(f"fibonacci(30): {fibonacci(30)}")
```

## 4. 创建自己的模块

```python
# 创建一个模块文件 mymath.py
module_code = '''
"""自定义数学模块"""

PI = 3.14159

def add(a, b):
    """加法"""
    return a + b

def multiply(a, b):
    """乘法"""
    return a * b

def circle_area(radius):
    """计算圆面积"""
    return PI * radius ** 2

if __name__ == "__main__":
    # 直接运行此文件时执行
    print("测试 mymath 模块")
    print(f"add(2, 3) = {add(2, 3)}")
    print(f"circle_area(5) = {circle_area(5)}")
'''

with open("mymath.py", "w", encoding="utf-8") as f:
    f.write(module_code)

print("mymath.py 模块已创建")
```

```python
# 使用自定义模块
import mymath

print(f"PI = {mymath.PI}")
print(f"add(5, 3) = {mymath.add(5, 3)}")
print(f"circle_area(10) = {mymath.circle_area(10)}")
```

## 5. 包（Package）

包是包含多个模块的目录。目录中必须有 `__init__.py` 文件。

```python
# 创建包结构
import os

# 创建包目录
os.makedirs("mypackage", exist_ok=True)

# 创建 __init__.py
init_code = '''
"""mypackage 包"""
from .utils import greet
from .calculator import add, subtract
'''

# 创建 utils.py
utils_code = '''
def greet(name):
    return f"Hello, {name}!"
'''

# 创建 calculator.py
calc_code = '''
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
'''

with open("mypackage/__init__.py", "w") as f:
    f.write(init_code)

with open("mypackage/utils.py", "w") as f:
    f.write(utils_code)

with open("mypackage/calculator.py", "w") as f:
    f.write(calc_code)

print("包已创建")
```

```python
# 使用包
from mypackage import greet, add

print(greet("World"))
print(f"add(10, 20) = {add(10, 20)}")
```
